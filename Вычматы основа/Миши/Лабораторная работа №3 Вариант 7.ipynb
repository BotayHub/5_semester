Метод Рунге-Кутты 3 порядка точности; $x\in (1, 1.2), \varepsilon = 10^{-4}, 2xyy'_{x}-y^{2}+5x=0, y(1)=1$

Выразим $y'_{x}$:

```python
def f(x,y):
    yder = ((y**2) - 5*x)/(2*x*y)
    return yder
```

```python
def f1(x,y):
    return f(x,y)
```

```python
def f2(x,y,h):
    return f(x + (h/3) , y + (h/3)*f1(x,y))
```

```python
def f3(x,y,h):
    return f(x + (2*h/3) , y + (2*h/3)*f2(x,y, h))
```

```python
def y_n1(x, y, h):
    y_n1 = y + (h/4)*(f1(x,y) + 3*f3(x,y,h))
    return y_n1
```

```python
a = 1 #начало отрезка
b = 1.2 #конец отрезка
N = 40 #начальное разбиение в 2h
y0 = 1
e = 10**(-4)

y2h_values = []
h = (b - a)/N

x_n = a 
y2h_values.append(y0)

for i in range(N):
    y_nplus1 = y_n1(x_n, y2h_values[i], h)
    y2h_values.append(y_nplus1)
    x_n += h

#увеличиваем разбиение в 2 раза:

N *= 2
yh_values = []
h = (b - a)/N

x_n = a 
yh_values.append(y0)

for i in range(N):
    y_nplus1 = y_n1(x_n, yh_values[i], h)
    yh_values.append(y_nplus1)
    x_n += h
```

```python
print(y2h_values)

print(yh_values)
```

Найдем разность значений в совпадающих узлах сетки:

```python
z = []
z.append(y0)
i = 0
while (i < N):
    z.append(yh_values[i+2])
    i+=2
#print(z)
```

```python
delta = []
i = 0
while i <= N/2:
    x = y2h_values[i] - z[i]
    delta.append(x)
    i += 1
print(delta)
```

```python
norm = max(delta)
print("delta max =", norm)
k = 3  #т.к. метод 3-го порядка
precision = norm/((2**k)-1)
if (precision <= e):
    print(precision) 
else: 
    print("error is bigger than epsilon")
```
